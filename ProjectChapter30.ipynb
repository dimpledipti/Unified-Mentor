##Data cleaning and preparation

```python
#Identifying missing values in a dataframe
df.isna().sum()
```

```python
#Identifying duplicate values from the dataframe
df.duplicated().sum()
```

```python
#Converting Attrition column into boolean for easy manipulation Yes=1 and No=0

df1 = df.replace({'Yes': 1, 'No': 0})
df1
```

```python
df1.shape
```

```python
#Converting dataframe into int or float datatype
df1_ = df1.select_dtypes(include=['number']).columns
df2 = df1[df1_].astype(int)
df2.info()
```

```python
df2.shape
```

```python
#Drop unnecessary column

df3 = df2.drop(columns = ['DailyRate', 'Education', 'EmployeeCount',
                          'EmployeeNumber', 'HourlyRate','MonthlyRate',
                          'NumCompaniesWorked', 'PercentSalaryHike',
                          'PerformanceRating','StandardHours','OverTime',
                          'StockOptionLevel', 'TotalWorkingYears',
                          'TrainingTimesLastYear','YearsWithCurrManager'])
df3.head()
```

##Data visualisation

```python
#Plotting heat map to find the dependencies of each column

plt.figure(figsize = (15,9))
sns.heatmap(df3.corr(),annot = True, linewidth = 1)
plt.title("Correlation")
plt.show()
```

```python
df4 = df3[df3['Attrition']==1]      # Positive attrition
df5 = df3[df3['Attrition']==0]      # Negative attrition
```

```python
# Relation between age and attrition

#first chart for Positive Attrition

sns.displot(data = df4, x = 'Age',  bins= 10,color='skyblue')
plt.title("Attrition Rate in various age group")
plt.xlabel("Positive Attrition Age")
plt.ylabel("count")
plt.show()

# #second chart for Negative attrition

# sns.displot(data = b, x = 'Age',  bins=10,color='skyblue')
# plt.xlabel("Negative Attrition Age group")
# plt.ylabel("count")
# plt.show()


```

```python
# Relation between 'Attrition' and 'MonthlyIncome'

#first chart for Positive Attrition

plt.figure(figsize = (10,5))
sns.barplot(data = df4, x = 'Age', y = 'MonthlyIncome',color='skyblue', width=0.8,legend='auto',errorbar = None)
plt.title("Monthly Income in various age group")
plt.xlabel('Age for Positive Attrition')
plt.ylabel("Monthly Income")
plt.xticks(rotation = 45)
plt.show()

#second chart for Negative attrition

plt.figure(figsize = (10,5))
sns.barplot(data = df5, x = 'Age', y = 'MonthlyIncome',color='skyblue', width=0.8,legend='auto',errorbar = None)
plt.xlabel("Age for Negative Attrition")
plt.ylabel("Monthly Income")
plt.xticks(rotation = 45)
plt.show()
```

```python
# plot relation between 'Years at a campany' and 'Income'

#first chart for Positive Attrition

plt.figure(figsize = (10,5))
sns.barplot(data = df4, x = 'YearsAtCompany', y = 'MonthlyIncome',color='skyblue', width=0.8,legend='auto',errorbar = None)
plt.title("Monthly Income vs YearsAtCompany")
plt.xlabel("Years At Company Positive Attrition")
plt.ylabel("Monthly Income")
plt.xticks(rotation = 45)
plt.show()

#second chart for Negative attrition

plt.figure(figsize = (10,5))
sns.barplot(data = df5, x = 'YearsAtCompany', y = 'MonthlyIncome',color='skyblue', width=0.8,legend='auto',errorbar = None)
plt.xlabel("Years At Company Negative Attrition")
plt.ylabel("Monthly Income")
plt.xticks(rotation = 45)
plt.show()
```
